### 1.1 Business Overview
Hypothyroidism is a medical condition where the thyroid gland fails to produce enough thyroid hormones, leading to symptoms like fatigue, weight gain, and depression. Common factors affecting hypothyroidism in the real world include iodine deficiency, autoimmune diseases (like Hashimoto's thyroiditis), certain medications, and radiation therapy.The frequency of hypothyroidism, goiters, and thyroid nodules increases with age. Hypothyroidism is most prevalent in elderly populations, with 2-20% of older age groups having some form of hypothyroidism. A Framingham study found hypothyroidism in 5.9% of women and 2.4% of men older than 60 years.Kenya Medical Research Institute aim to identify patterns and relations between demographic factors, medical history, and thyroid function test results. By analyzing these patterns, they hope to improve diagnostic accuracy, tailor treatments more effectively, and enhance the overall understanding of hypothyroidis management.

```python
# importing necessary libraries
import pandas as pd
import pprint
import re
```

```python
class uncleanAnalysis:
    def __init__(self, path):
      # load dataset
        self.df = pd.read_csv(path)

    def basic_info(self):
        print("# First five rows of the dataset")
        # first 5 entries
        pprint.pp(self.df.head())

        # shape of hypothyroid dataset
        print("\n# Shape of the dataset")
        pprint.pp(self.df.shape)

        # dataset columns
        print("\n# Columns of the dataset")
        print(self.df.columns)

        # first 10 unique values within the columns
        print("\n# Distinct categories or values in the dataset")
        for column in self.df.columns:
            pprint.pp(f"{column}: {self.df[column].unique()[:10]}")

        # More info abt the dataset
        print("\n# Info of the dataset")
        pprint.pp(self.df.info())

        # mean, percentiles, count, etc
        print("\n# Statistical summary before cleaning")
        pprint.pp(self.df.describe())

        # null values
        print("\n# Checking for missing values")
        pprint.pp(self.df.isnull().sum())

        # column data typpes
        print("\n# Data types of columns")
        pprint.pp(self.df.dtypes)

# import file here
path = "sample_data/hypothyroid.csv"

# Assign class to variable 'analysis'
analysis = uncleanAnalysis(path)

# Call the class
analysis.basic_info()
```

```python
class cleanAnalysis(uncleanAnalysis):
    def __init__(self, path):
        super().__init__(path)
        self.clean_data()

    def clean_data(self):
        # Fill "?" with modal values
        for column in self.df.columns:
            mode = self.df[column].mode()[0]
            self.df[column] = self.df[column].replace('?', mode)

        # Define a function to clean numeric values
        def clean_numeric(x):
            return re.sub(r'[^\d.]+', '', str(x))

        # Clean and convert appropriate columns to numeric
        numeric_cols = ['age', 'T3', 'TT4', 'T4U', 'FTI']
        for col in numeric_cols:
            self.df[col] = self.df[col].map(clean_numeric)
            self.df[col] = pd.to_numeric(self.df[col], errors='ignore')

        # Convert binary columns to boolean
        bool_cols = [
            'on_thyroxine', 'query_on_thyroxine', 'on_antithyroid_medication',
            'thyroid_surgery', 'query_hypothyroid', 'query_hyperthyroid',
            'pregnant', 'sick', 'tumor', 'lithium', 'goitre', 'TSH_measured',
            'T3_measured', 'TT4_measured', 'T4U_measured', 'FTI_measured', 'TBG_measured'
        ]
        for col in bool_cols:
            self.df[col] = self.df[col].map({'t': True, 'f': False, 'y': True, 'n': False})

        # Map M and F to Male and Female
        self.df['sex'] = self.df['sex'].map({'M': 'Male', 'F': 'Female', 'm': 'Male', 'f': 'Female'})
        
        # Fill remaining missing values
        for col in self.df.columns:
            if self.df[col].dtype == "object":
                self.df[col].fillna(self.df[col].mode()[0], inplace=True)
            else:
                self.df[col].fillna(self.df[col].mean(), inplace=True)

        # Print to verify the cleaning steps
        print("\n# Data types after cleaning")
        pprint.pprint(self.df.dtypes)

        print("\n# Checking for missing values after cleaning")
        pprint.pprint(self.df.isnull().sum())


# Assign class to variable 'clean_analysis'
clean_analysis = cleanAnalysis(path)

# Call the class to perform cleaning and basic info
clean_analysis.basic_info()
```
